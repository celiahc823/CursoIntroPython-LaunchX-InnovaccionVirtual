### Desactivar un entorno virtual

Se ha creado creado un entorno virtual y le has agregado un paquete. Sin embargo, es posible que estés trabajando en varios proyectos de Python y necesites cambiar entre ellos. Para hacer eso, debes salir (desactivar) tu entorno virtual.

Ejecuta el comando ``deactivate``:

<img src="kata_a6.jpg">

Listo. Quedó desactivado el entorno virtual.
